```python
import pandas as pd
```

```python
df_securities = pd.read_csv('Securities.csv')
df_securities
```

```python
for col in list(df_securities.columns):
    first_valid_index = df_securities[col].first_valid_index()
    if first_valid_index > 0:
        df_securities[col][0:first_valid_index-1] = 'Null'
    
df_securities.fillna(method='ffill', inplace=True)
df_securities['Date'] = pd.to_datetime(df_securities['Date']).dt.date
```

```python
main_df = pd.read_csv('macroeconomic_indicators.csv')
main_df['Date'] = pd.to_datetime(main_df['Date']).dt.date
```

```python
final_df = main_df.merge(df_securities, how='left', left_on = 'Date', right_on = 'Date')
final_df.fillna(method='ffill', inplace=True)
final_df.replace('Null', 0).to_csv('final_output.csv', index=False)

final_df
```
